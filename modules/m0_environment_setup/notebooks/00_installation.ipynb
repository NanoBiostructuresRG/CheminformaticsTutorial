### Contents

This notebook introduces the first step in this long journey. Before diving into the API, we must ensure your computational environment is correctly configured. This section covers the deployment of the necessary tools to execute chemoinformatics workflows locally or in the cloud.

The activities are structured to guide you through the process of:

* Platform Selection: Choosing your workspace between **Google Colab** (cloud-based) or a local distribution using **Anaconda** and Miniconda.
* Interface Configuration: Setting up the development environment using **Jupyter Notebook**, **JupyterLab**, or **VS Code** to manage `.ipynb` files.
* Course Environment Creation: Building a dedicated Python environment with the specific libraries (Requests, Pandas, RDKit) required for chemical data handling.
* Dependency Management: Verifying installations to ensure seamless communication between your local machine and web-based chemical databases.

This setup is designed to be completed in approximately **30–60 minutes**.

Having a robust and isolated environment is the foundation of reproducible science. A well-configured workspace ensures that the code you write today will remain functional and portable across different platforms.

## Google Colab

Google Colaboratory (or **"Colab"**) is a cloud-based service that allows you to execute Python code through your browser. For this course, it serves as our Zero-Configuration Laboratory, lowering the technical barrier for beginners and ensuring a standardized workflow.

This platform is specifically chosen because it addresses key challenges in computational research:

- **No Local Installation Required:** You can start querying PubChem immediately without resolving complex package dependency issues or local Python conflicts.
- **Cloud-Based Resources:** Colab provides free access to hosted RAM and GPUs, allowing data-intensive chemical tasks to run independently of your local hardware.
- **Preinstalled Chemoinformatics Ecosystem:** Core libraries like Pandas and Requests are ready to use, while specialized tools like RDKit can be integrated with a single command.
- **Seamless Collaboration:** Integration with Google Drive allows you to save your progress, share your findings, and collaborate on your research notebooks in real-time.


Based on these reasons, we think that Google Colab is the ideal starting point for this course, as it ensures a unified environment for all students, eliminating the "it works on my machine" hurdles during the first stages of data retrieval.



## Anaconda

While cloud platforms are excellent for testing, professional research requires a robust local setup. Anaconda is the industry-standard distribution for scientific computing, providing the tools necessary to manage complex chemical software dependencies.

In this section, you will learn to transition to a professional local workflow by:

- **Building Reproducible Environments:** Learning to create isolated "virtual spaces" where specific versions of libraries (like RDKit or Pandas) coexist without conflicts.
- **Mastering Package Management:** Using the conda manager to reliably install complex scientific tools that are often difficult to configure manually.
- **Integrating Professional Tooling:** Setting up a workspace that connects JupyterLab and VS Code, supporting both interactive data exploration and script-based automation.
- **Local Data Control:** Managing large datasets and project outputs directly on your file system for faster access and better organization.
- **Adopting Industry Standards:** Emulating real-world computational chemistry practices used in both academic research and the pharmaceutical industry.


## Jupyter Notebook & JupyterLab

The Jupyter ecosystem is the gold standard for data storytelling in science. It allows you to combine live code, chemical visualizations, and rich narrative text into a single, shareable research document.

You will learn to navigate these interactive environments by:

- **Mastering the Classic Notebook:** Utilizing a linear interface ideal for step-by-step tutorials and guided chemoinformatics workflows.
- **Transitioning to JupyterLab:** Managing an advanced workspace that integrates multiple notebooks, file browsers, and terminals in a single tabbed interface.
- **Understanding Interactive Execution:** Learning how the Kernel maintains your variables in memory, enabling dynamic, iterative analysis without re-running the entire script.
- **Standardizing Scientific Documentation:** Using **Markdown** to transform your raw code into a professional, reproducible computational report.


**What is the main difference?:**

The primary distinction lies in the interface architecture. While the classic **Jupyter Notebook** is designed for a single-document, linear flow (one file per tab), **JupyterLab** is a full "Integrated Development Environment" (IDE). It allows you to open multiple notebooks, CSV files, and terminals side-by-side in a flexible, modular layout, making it much more efficient for complex research projects that require multitasking.

Remember that Jupyter notebooks function as your digital laboratory notebooks, where every data retrieval and transformation is recorded in a transparent and reproducible manner.

## Visual Studio Code

Visual Studio Code (**VS Code**) is a lightweight but powerful code editor that serves as the central hub of this course. It is designed for researchers who need to transition from interactive exploration to professional, scalable development.

In this section, you will learn to manage your chemoinformatics projects by:

- **Integrating a Unified Workspace:** Discovering how to execute `.ipynb` notebooks and traditional Python scripts side-by-side within a single, streamlined interface.
- **Leveraging the Extension Ecosystem:** Setting up official Python and Jupyter tools to unlock intelligent code completion (IntelliSense) and real-time error highlighting.
- **Managing Environments Locally:** Learning how VS Code detects and connects to your Conda environments, ensuring your code always runs with the correct library versions.
- **Git & GitHub Readiness:** Preparing your workspace for version control, a critical skill for collaborating on large-scale computational projects.

**Why use VS Code over standard Jupyter?:**

The main advantage is versatility. VS Code allows you to manage your entire project—notebooks, raw data files, documentation, and terminal—in a single window. It is designed for developers who need more power, better organization, and the ability to turn experimental notebook code into reusable software libraries.

## Create the Course Environment

The final step in the setup process is to bring all these tools together into a single, reproducible computational environment. In practical terms, we will create a dedicated workspace specifically configured for this course.

This environment ensures that you have:

- **Isolated Dependencies:** Your course setup remains independent from other Python projects and system-level package changes.
- **Verified Scientific Libraries:** Core libraries such as **requests**, **pandas**, and **RDKit** are installed in compatible versions that support the cheminformatics workflows developed throughout the course.
- **Reproducibility:** A standardized environment can be recreated across different machines, ensuring that analyses and results remain consistent.
- **Workflow Stability:** By working within a dedicated environment, you reduce installation conflicts and create a more reliable foundation for future modules.

### Step 1 — Install Anaconda

Download and install the **Anaconda Distribution** (compatible with your operating system) from:

https://www.anaconda.com/download

Choose the installer for your operating system (Windows / macOS / Linux) and follow the setup instructions.

After installation, open the **Anaconda Prompt** (Windows) or a terminal (macOS/Linux) and verify:

```bash
conda --version
```

If a version number appears, Anaconda is correctly installed.



### Step 2 — Launch Jupyter Notebook / JupyterLab

With Anaconda installed, you can start an interactive computing session using Jupyter.

To launch JupyterLab (recommended), open the **Anaconda Prompt** (Windows) or a terminal (macOS/Linux) and run:

```bash
jupyter lab
```

Alternatively, you can launch the classic Jupyter Notebook interface:

```bash
jupyter notebook
```

A new browser window will open automatically.

**Your First Notebook:**

- Navigate to your project folder
- Click "New" → "Notebook" -> "Python 3"
- Rename the notebook (e.g., intro_notebook.ipynb)
-You are now ready to execute code interactively

Run the following code in a new cell to verify your environment:

```bash
import pandas as pd
print("Jupyter is working correctly")
```

If the cell executes without errors, your Jupyter environment is ready.

>**Note:** Jupyter notebooks combine code, output, and documentation in a single file.
> JupyterLab provides a more modern interface with tabs and file navigation.
> This environment will be used extensively in Module 2. 

### Step 3 — Install VS Code

Download and install VS Code from:

https://code.visualstudio.com/

### Step 4 — Create the Course Environment

From the root directory of this repository, run:

```bash
conda env create -f m0_environment_setup/environment.yml
```

Once the installation completes, activate the environment:

```bash
conda activate cheminf_tutorial
```

You should now see `(cheminf_tutorial)` at the beginning of your terminal line.


### Step 5 — Configure VS Code to Use the Environment

- Open the repository folder in VS Code
- Press `Ctrl + Shift + P`
- Select: Python: **Select Interpreter**
- Choose `cheminf_tutorial`


### Step 6 — Register the Environment in Jupyter (Optional but Recommended)

```bash
python -m ipykernel install --user --name cheminf_tutorial --display-name "Python (cheminf_tutorial)"
```

This allows you to select the environment inside Jupyter notebooks.

### Step 7 — Verify RDKit Installation

Create a new notebook and run:

```bash
from rdkit import Chem
Chem.MolFromSmiles("CCO")
```

If no errors appear, your environment is correctly configured.
